## **Librerías**

```python
import matplotlib.pyplot as plt   # Librería para graficar
import numpy as np                # Librería para manejo de vectores y utilidades matemáticas
```

## **Funciones algebraicas**

### **Función lineal**

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Rango_{f} = (-\infty, \infty)$

```python
def funcion(m,x,b):
  return m*x + b

resolucion = 100  # Número de puntos
m = 10
b = 5
x = np.linspace(-10.0, 10.0, num=resolucion) # Genera una serie de puntos en un rango
y = funcion(m,x,b)

# Graficar
fig, ax = plt.subplots()
ax.plot(x,y)   # Hasta aquí ya se puede ver una gráfica
ax.grid()
ax.axhline(y=0, color="r")  # Dibuja la linea horizontal con y
ax.axvline(x=0, color="r")  # Dibuja la linea vertical con x
```

### Función Polinómicas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una función que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremoc coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinómio de grado 7.


### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polinómicas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$

```python
def funcion(x):
  return x**2

resolucion = 100  # Número de puntos
x = np.linspace(-10.0, 10.0, num=resolucion) # Genera una serie de puntos en un rango
y = funcion(x)

# Graficar
fig, ax = plt.subplots()
ax.plot(x,y)   # Hasta aquí ya se puede ver una gráfica
ax.grid()
ax.axhline(y=0, color="r")  # Dibuja la linea horizontal con y
ax.axvline(x=0, color="r")  # Dibuja la linea vertical con x
```

```python
def funcion(x):
  return 2*x**7 - x**4 + 3*x**2 + 4

resolucion = 100  # Número de puntos
x = np.linspace(-5.0, 5.0, num=resolucion) # Genera una serie de puntos en un rango
y = funcion(x)

# Graficar
fig, ax = plt.subplots()
plt.xlim(-1.5,1.5)  # Establece limites del eje x
plt.ylim(-1,8)  # Establece limites del eje y
ax.plot(x,y)   # Hasta aquí ya se puede ver una gráfica
ax.grid()
ax.axhline(y=0, color="r")  # Dibuja la linea horizontal con y
ax.axvline(x=0, color="r")  # Dibuja la linea vertical con x
```

Lo que se observa es que por lo menos con la función `linspace` de `numpy` se establecen una serie de numeros que estan limitados inicialmente por los valores de limite inicial y limite final, entre estos dos se hace una división de acuerdo al número de puntos que queremos o de valores `x`, osea que entre ménos diferencia haya entre cada `x` generado más suave se vuelve la linea de la gráfica por lo que tiene un efecto más curvo.







# Funciones trascendentes

Son funciones que no pueden ser expresadas con polinomios.

```python
def funcion(x):
  return np.cos(x)

resolucion = 100  # Número de puntos
x = np.linspace(-10.0, 10.0, num=resolucion) # Genera una serie de puntos en un rango
y = funcion(x)

# Graficar
fig, ax = plt.subplots()

ax.plot(x,y)   # Hasta aquí ya se puede ver una gráfica
ax.grid()
ax.axhline(y=0, color="r")  # Dibuja la linea horizontal con y
ax.axvline(x=0, color="r")  # Dibuja la linea vertical con x
```

### Función exponencial

Tienen la forma de $$f(x)=a^x$$ donde la base $a$ es una constante positiva. Un gran ejemplo de una función exponencial es usando la base como el número de euler:

$$f(x)=e^x$$

```python
def funcion(x):
  return 2**x

resolucion = 100  # Número de puntos
x = np.linspace(-1.0, 1.0, num=resolucion) # Genera una serie de puntos en un rango
y = funcion(x)

# Graficar
fig, ax = plt.subplots()

ax.plot(x,y)   # Hasta aquí ya se puede ver una gráfica
ax.grid()
ax.axhline(y=0, color="r")  # Dibuja la linea horizontal con y
ax.axvline(x=0, color="r")  # Dibuja la linea vertical con x
```

### Función logaritmo

El logaritmo está definido por la **realación**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que está elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$

```python
def funcion(x):
  return np.log2(x)

resolucion = 100  # Número de puntos
x = np.linspace(0.1, 256.0, num=resolucion) # Genera una serie de puntos en un rango
y = funcion(x)

# Graficar
fig, ax = plt.subplots()

ax.plot(x,y)   # Hasta aquí ya se puede ver una gráfica
ax.grid()
ax.axhline(y=0, color="r")  # Dibuja la linea horizontal con y
ax.axvline(x=0, color="r")  # Dibuja la linea vertical con x
```

# Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def heaviside(x):
  y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x >= 0:
      y[idx] = 1.0
  return y

resolucion = 100  # Número de puntos
x = np.linspace(-10.0, 10.0, num=resolucion) # Genera una serie de puntos en un rango
y = heaviside(x)

# Graficar
fig, ax = plt.subplots()

ax.plot(x,y)   # Hasta aquí ya se puede ver una gráfica
ax.grid()
ax.axhline(y=0, color="r")  # Dibuja la linea horizontal con y
ax.axvline(x=0, color="r")  # Dibuja la linea vertical con x
```
